```python
import pandas as pd
file_paths = [
    "../data/raw/api_data_aadhar_demographic_0_500000.csv",
    "../data/raw/api_data_aadhar_demographic_500000_1000000.csv",
    "../data/raw/api_data_aadhar_demographic_1000000_1500000.csv",
    "../data/raw/api_data_aadhar_demographic_1500000_2000000.csv",
    "../data/raw/api_data_aadhar_demographic_2000000_2071700.csv"
]
df_list = []

for file in file_paths:
    temp_df = pd.read_csv(file)
    
    # Standardize column names
    temp_df.columns = temp_df.columns.str.strip().str.lower()
    
    # Optional: keep track of source file
    temp_df['source_file'] = file.split('/')[-1]
    
    df_list.append(temp_df)

combined_df = pd.concat(df_list, ignore_index=True)
combined_df.shape
combined_df.head()
combined_df.tail()
combined_df.duplicated().sum()
combined_df.to_csv(
    "../data/processed/aadhaar_demographic_all_records.csv",
    index=False
)
```
